```python
import numpy as np

# Metadata for clarity
SUBJECTS = ["AI", "Python", "Data Analysis"]
TIMES = ["Morning", "Afternoon", "Evening"]

data = {
     "Banan":   {"interest": np.array([9,8,2]), "time": np.array([1,0,0])},
    "Ibtihal": {"interest": np.array([9,9,1]), "time": np.array([1,0,0])},
    "Raghad":  {"interest": np.array([7,1,9]), "time": np.array([0,0,1])},
    "Ramah":   {"interest": np.array([3,2,9]), "time": np.array([0,0,1])},
    "Abrar":    {"interest": np.array([8,7,3]), "time": np.array([1,0,0])},
    "Alaa":   {"interest": np.array([2,2,8]), "time": np.array([0,1,0])},
    "Bayan":   {"interest": np.array([1,10,2]), "time": np.array([0,1,0])},
    "Khulud":    {"interest": np.array([8,6,9]), "time": np.array([0,0,1])},
    "Leen":    {"interest": np.array([8,2,8]), "time": np.array([0,1,0])},
    "Reema":    {"interest": np.array([3,9,3]), "time": np.array([0,1,0])},
    "Lana":    {"interest": np.array([1,2,10]), "time": np.array([0,0,1])},
    "Layan":    {"interest": np.array([8,2,4]), "time": np.array([1,0,1])}
}

def cosine(v1, v2):
    denom = (np.linalg.norm(v1) * np.linalg.norm(v2))
    return np.dot(v1, v2) / denom if denom != 0 else 0

# 1. Similarity Calculations
all_pairs = []
students = list(data.keys())
for i in range(len(students)):
    for j in range(i + 1, len(students)):
        s1, s2 = students[i], students[j]
        int_sim = cosine(data[s1]["interest"], data[s2]["interest"])
        t_overlap = np.dot(data[s1]["time"], data[s2]["time"])
        score = 0.6 * int_sim + 0.4 * t_overlap
        all_pairs.append(((s1, s2), score, int_sim, t_overlap))

# 2. Ranking and Matching
all_pairs.sort(key=lambda x: x[1], reverse=True)
final_matches = []
used = set()

for pair_info in all_pairs:
    (s1, s2), score, i_sim, t_over = pair_info
    if s1 not in used and s2 not in used:
        final_matches.append(pair_info)
        used.add(s1)
        used.add(s2)

# 3. Enhanced Table Output
print("\n" + "="*85)
print(f"{'PARTNER MATCHING SYSTEM':^85}")
print("="*85)

# Table Header
header = f"{'Pairs':<6} | {'Partner 1':<12} | {'Partner 2':<12} | {'Shared Interest':<15} | {'Time':<10} | {'Match %'}"
print(header)
print("-" * 85)

# Table Rows
for i, ((s1, s2), score, i_sim, t_over) in enumerate(final_matches, 1):

    # Identify common interest (argmax of sum)
    common_interest_idx = np.argmax(data[s1]["interest"] + data[s2]["interest"])
    m_name = SUBJECTS[common_interest_idx]

    # Identify time slot (argmax of intersection)
    time_idx = np.argmax(data[s1]["time"] * data[s2]["time"])
    t_name = TIMES[time_idx] if t_over > 0 else "N/A"

    # Formatting each row
    row = f"{i:<4} | {s1:<12} | {s2:<12} | {m_name:<15} | {t_name:<10} | {score*100:>6.1f}%"
    print(row)

print("="*85)
```
